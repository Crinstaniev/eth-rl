```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_stable_alpha = pd.read_csv("results/stable_alpha.csv")
df_a2c = pd.read_csv("results/a2c.csv")
df_ddpg = pd.read_csv("results/ddpg.csv")
```

```python
df_stable_alpha.head()
```

```python
df_a2c.head()
```

```python
df_ddpg.head()
```

```python
# compare honest proportion using plot
plt.plot(df_stable_alpha["honest_proportion"], label="stable alpha")
plt.plot(df_a2c["honest_proportion"], label="a2c")
plt.plot(df_ddpg["honest_proportion"], label="ddpg")
plt.legend()
plt.show()
```

```python
# compare alpha using plot
plt.plot(df_stable_alpha["alpha"], label="stable alpha")
plt.plot(df_a2c["alpha"], label="a2c")
plt.plot(df_ddpg["alpha"], label="ddpg")
plt.legend()
plt.show()
```

```python
# compare sum of active balance
plt.plot(df_stable_alpha["sum_of_effective_balance"], label="stable alpha")
plt.plot(df_a2c["sum_of_effective_balance"], label="a2c")
plt.plot(df_ddpg["sum_of_effective_balance"], label="ddpg")
plt.legend()
plt.show()
```
